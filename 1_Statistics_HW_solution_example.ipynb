## Стратификация

Начнем со следующего примера:

Предположим, что нам нужно оценить среднее число голосов, поданных за каждого кандидата на выборах. Предположим, что в стране 3 города: в городе А живёт 1 миллион заводских рабочих, в городе Б живёт 2 миллиона офисных работников, а в городе В живёт 3 миллиона пенсионеров. Мы можем выбрать получение случайной выборки размером в 60 голосов со всей популяции, но есть некоторая вероятность, что случайная выборка окажется плохо сбалансирована по этим городам и, следовательно, будет необъективна и малополезна («средняя температура по больнице»), вызывая значительную погрешность в оценке. Взамен этого, если мы выберем использовать простую случайную выборку в 10, 20 и 30 голосов из городов А, Б и В соответственно, мы можем получить меньшую погрешность в оценке при том же общем размере выборки. Такой прием называется стратификацией.

Подробнее о стратификации [здесь](https://compscicenter.ru/media/slides/math_stat_2017_spring/2017_02_27_math_stat_2017_spring_qWLJs8i.pdf)

### Задача

Предположим, что генеральная совокупность - смесь 3х нормально распределенных СВ. Иными словами, генеральную совокупность можно разбить на 3 страты.
$$
    F(X) = a_1 F(X_1) + a_2 F(X_2) + a_3 F(X_3)
$$

** Требуется:**
* Вывести формулы и подсчитать точечные оценки математического ожидания и дисперсии для среднего значения подвыборок, образованных разными способами:
    * случайный выбор из всей генеральной совокупности
    * случайный выбор из страт пропорционально долям
    * случайный выбор согласно оптимальному разбиению на подсэмплы
Оценки вычисляются из результатов повторения эксперимента 1000 раз

Эксперименты провести для 3ех разных случаев:
* все страты имеют одинаковые мат. ожидания и дисперсии
* страты имеют разные мат.ожидания, но одинаковые дисперсии
* страты имеют разные и мат. ожидания и дисперсии.

Также провести эксперимент для разных размеров генеральной совокупности: 40 и 500. Параметры нормальных распределений, доли и размер подвыборки задайте самостоятельно

## Решение Задания 2

### Вывод формул:

$x_1, x_2, ..., x_N$ - генеральная совокупность (ГС) конечного объема $N$.

Математическое ожидание ГС:

$$
\mu = \frac{1}{N} \sum_{i=1}^N x_i
$$

Пусть элементы генеральной совокупности выбираются случайным образом <b>с возвращением</b>

$(X_1, X_2, ..., X_n)$ - выборка из ГС

$n$ - объем выборки

Точечная <b>оценка мат. ожидания</b> ГС - это выборочное среднее:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^n X_i
$$

Математическое ожидание $\bar{X}$:

$$
E\bar{X} = \frac{1}{n} \sum_{i=1}^n EX_i = \mu
$$

Дисперсия $\bar{X}$:

$$
Var(\bar{X}) = \frac{1}{n^2} \sum_{i=1}^n \sum_{j=1}^n Cov(X_i, X_j)
$$

Т.к. элементы выборки извлекаются с возвращением, то $Cov(X_i, X_j) = 0, i \ne j$. Тогда:

$$
Var(\bar{X}) = \frac{1}{n^2} \sum_{i=1}^n Var(X_i) = \frac{\sigma^2}{n}
$$

#### Стратифицированные выборки

В рамках страты $k$ берется выборка объема $n_k$

$(X_{1k}, X_{2k}, ..., X_{n_k k}), k = \bar{1,L}$

Тогда выборочное среднее в рамках $k$-ой страты:

$$
\bar{X_k} = \frac{1}{n_k} \sum_{i=1}^{n_k} X_{ik}
$$

Оценка мат. ожидания смеси страт:

$$
\mu = \sum_{k=1}^L W_k \mu_k \Rightarrow \bar{X_S} = \sum_{k=1}^L W_k \bar{X_k}
$$

$W_k$ - доля $k$-ой страты

Мат. ожидание $\bar{X_S}$:

$$
E\bar{X_S} = \sum_{k=1}^L W_k E\bar{X_k} = \sum_{k=1}^L W_k \mu_k
$$

Дисперсия $\bar{X_S}$:

$$
D(\bar{X_S}) = \sum_{k=1}^L {W_k}^2 D (\bar{X_k}) = \sum_{k=1}^L {W_k}^2 \frac{{\sigma_k}^2}{n_k}
$$

#### Объемы выборок при пропорциональном сэмплировании:

$$
n_k = n W_k, k = \bar{1,L}, n = n_1 + n_2 + ... + n_L
$$

#### Объемы выборок при оптимальном сэмплировании (Неймана):

$$
n_k = \tilde{n_k} = n \frac{W_k \sigma_k}{\sum_{k=1}^L W_k \sigma_k} , k = \bar{1,L}, n = n_1 + n_2 + ... + n_L
$$

````python
class Experiment:

    def __init__(self, means, stds, random_state=1):
        """Инициализирует наш эксперимент и сохраняем заданные распределения
        
        :param means: Список матожиданий для нормальных распределний
        :param stds: Список стандартных отклонений для нормальных распределний
        :param random_state: Параметр фиксирующий случайность. Нужен для того, чтобы при воспроизведении
        эксперимента повторно с теми же входными параметрами, результаты оставались такими же
        """
        self.strats = [st.norm(mean, std) for mean, std in zip(means, stds)]
    
    def sample(self, sizes):
        """Создает выборку генеральной совокупности
        
        :param sizes: Список с размерами выборок соответсвующих им нормальных распределений
        """
        self.strats_samples = [rv.rvs(size) for rv, size in zip(self.strats, sizes)]
        self.general_samples = np.hstack(self.strats_samples)
        self.N = self.general_samples.shape[0]
        
        # количество страт
        self.count_strats = len(sizes)
        
        # доли каждой страты в ГС
        self.ws = [size/self.N for size in sizes]
        
        #мат ожидание и стандартное отклонение выборки ГС
        self.m = np.mean(self.general_samples)
        self.sigma = np.std(self.general_samples)
        
        # мат ожидания страт и стандартные отклонения страт в выборке ГС
        self.ms = [np.mean(strat_sample) for strat_sample in self.strats_samples]
        self.sigmas = [np.std(strat_sample) for strat_sample in self.strats_samples]
        
        return (self.m, self.sigma)
        
    def random_subsampling(self, size):
        """Создает случайную подвыборку из генеральной совокупности
        
        :param sizes: размер подвыборки
        """
        rc = np.random.choice(self.general_samples, size=size)
        
        return rc
        
    def proportional_subsampling(self, size):
        """Создает подвыборку с кол-вом элементов, пропроциональным долям страт
        
        :param sizes: размер подвыборки
        """
        # Вычисляем пропорциональный размер страт
        self.strats_size_proport = [int(np.floor(size*w)) for w in self.ws]
        
        rc = []
        for k in range(len(self.strats_size_proport)):
            rc.append(np.random.choice(self.strats_samples[k],
                                       size=self.strats_size_proport[k]))
        
        return rc
        
    def optimal_subsampling(self, size):
        """Создает подвыборку с оптимальным кол-вом элементов относительно страт
        
        :param sizes: размер подвыборки
        """
        # Вычисляем оптимальный размер страт
        # сумма в знаменателе
        sum_denom = 0
        for k in range(self.count_strats):
            sum_denom += self.ws[k] * self.sigmas[k]
        
        self.strats_size_optimal = [int(np.floor((size*w*sigma)/sum_denom)) for w, sigma in zip(self.ws, self.sigmas)]
        if 0 in self.strats_size_optimal:
            raise ValueError('Strats size is 0, please change variance of smallest strat!')
        
        rc = []
        for k in range(len(self.strats_size_optimal)):
            rc.append(np.random.choice(self.strats_samples[k],
                                       size=self.strats_size_optimal[k]))
        
        return rc
    
    def run_experiments(self, subsampling_method, n_experiments=1000):
        """Проводит серию экпериментов и сохраняет результаты
        
        :param subsampling_method: метод для создания подвыборки
        :param n_experiments: кол-во запусков эксперимента
        """
        means_s = []
        
        n_sub = 100
        if(len(self.general_samples)<100):
            n_sub = 20
        
        if(subsampling_method == 'random_subsampling'):
            for n in range(n_experiments):
                rc = self.random_subsampling(n_sub)
                
                mean = rc.sum()/len(rc)
                means_s.append(mean)
                
        else:
            for n in range(n_experiments):
                if(subsampling_method == 'proportional_subsampling'):
                    rc = self.proportional_subsampling(n_sub)
                elif(subsampling_method == 'optimal_subsampling'):
                    rc = self.optimal_subsampling(n_sub)
                
                strats_mean = []
        
                for k in range(len(rc)):
                    strats_mean.append(sum(rc[k])/len(rc[k]))
        
                # Расчет среднего для смеси
                means_s.append(sum([w_k*mean_k for w_k, mean_k in zip(self.ws, strats_mean)]))
                
        return means_s
````

````python
means_random = []
means_proport = []
means_optimal = []

def show_experiment_results(experiment, N):
    rcParams['figure.figsize'] = 5, 3
    plt.hist(experiment.general_samples, bins=15)
    title = 'Плотность выборки ГС (N='+str(N)+')'
    plt.title(title)
    plt.show()
    
    means_random = experiment.run_experiments('random_subsampling')
    means_proport = experiment.run_experiments('proportional_subsampling')
    try:
        means_optimal = experiment.run_experiments('optimal_subsampling')
    except ValueError as e:
        print(str(e))
        
    means_lists = pd.DataFrame(
        {'1 random subsampling': means_random,
         '2 proportional subsampling': means_proport,
         '3 optimal subsampling': means_optimal
        })
    
    rcParams['figure.figsize'] = 7, 6
    sns.boxplot(data=means_lists, width=0.3)
    plt.title('Распределение среднего подвыборки для трех способов сэмплирования')
    plt.show();
    
    print('Оценки среднего подвыборки')
    print('\nСлучайный выбор из всей ГС:')
    print(str.format('\tтеоретическое значение мат ожидания: {0:.3f}', experiment.m))
    print(str.format('\tэмпирическое значение мат ожидания: {0:.3f}', np.mean(means_random)))
    print(str.format('\n\tтеоретическое значение дисперсии: {0:.3f}', (experiment.sigma**2)/n))
    print(str.format('\tэмпирическое значение дисперсии: {0:.3f}', np.var(means_random)))
    
    m_s = sum([m_k*w_k for m_k, w_k in zip(experiment.ms, experiment.ws)])
    d_s_proport = sum([(w_k**2)*((sigma_k**2)/n_k)
                       for n_k, w_k, sigma_k
                       in zip(experiment.strats_size_proport, experiment.ws, experiment.sigmas)])
    print('\n\nСлучайный выбор из страт пропорционально долям:')
    print(str.format('\tтеоретическое значение мат ожидания: {0:.3f}', m_s))
    print(str.format('\tэмпирическое значение мат ожидания: {0:.3f}', np.mean(means_proport)))
    print(str.format('\n\tтеоретическое значение дисперсии: {0:.3f}', d_s_proport))
    print(str.format('\tэмпирическое значение дисперсии: {0:.3f}', np.var(means_proport)))
    
    d_s_optimal = sum([(w_k**2)*((sigma_k**2)/n_k)
                       for n_k, w_k, sigma_k
                       in zip(experiment.strats_size_optimal, experiment.ws, experiment.sigmas)])
    print('\n\nСлучайный выбор из страт согласно оптимальному разделению:')
    print(str.format('\tтеоретическое значение мат ожидания: {0:.3f}', m_s))
    print(str.format('\tэмпирическое значение мат ожидания: {0:.3f}', np.mean(means_optimal)))
    print(str.format('\n\tтеоретическое значение дисперсии: {0:.3f}', d_s_optimal))
    print(str.format('\tэмпирическое значение дисперсии: {0:.3f}', np.var(means_optimal)))
````

## Последовательный критерий Вальда (критерий [SPRT](https://en.wikipedia.org/wiki/Sequential_probability_ratio_test))

[Видео с объяснением](https://www.youtube.com/watch?v=4NYbTUIkBbU&index=4&list=PLlb7e2G7aSpRG_Ve6kuxc9Biuwyf42kJx) (начало 1:06:20) 

### Задача

* Примените к сгенерированным данным последовательный критерий Вальда силы $(\alpha, \beta)$ (критерий Вальда должен быть реализован самостоятельно). Для этого:
    * Задайте нулевую и альтернативную гипотезы о равенствах матожиданий конкретным величинам
    * Задайте $\alpha, \beta$
    * Выпишите формулы аппроксимации границы
    * Провести эксперимент
    * Результат эксперимента отобразить на графике
* Проанализируйте поведение критерия при разных долях компонент в смеси
* Какой объем выборок потребовался в каждом случае?
* Выпишите нижнюю оценку для среднего числа наблюдений при заданных $(\alpha, \beta)$ в случае справедливости нулевой и альтернативной гипотез

## Решение Задания 3

Гипотезы:

$H_0: a = a_0$

$H_1: a = a_1$

Возьмем вероятность ошибки первого рода равной:
$\alpha = 0.04$

А вероятность ошибки второго рода равной:
$\beta = 0.001$

Ищем подходящие $c_0, c_1$.

Пусть $0 < c_0 < 1 < c_1 < \infty$

Тогда:
$$
\alpha \leqslant \frac{1-\beta}{c_1} \leqslant \frac{1}{c_1}
$$

$$
\beta \leqslant (1-\alpha)c_0 \leqslant c_0
$$

<b>Аппроксимация границ $c_0, c_1$</b>

Введем значения ${c_0}', {c_1}'$

$$
{c_0}' = \frac{\beta}{1 - \alpha}
$$

$$
{c_1}' = \frac{1-\beta}{\alpha}
$$

$$
\alpha' \leqslant \frac{\alpha}{1-\beta}
$$

$$
\beta' \leqslant \frac{\beta}{1-\alpha}
$$

Если $\alpha$ и $\beta$ невелики (от 0.001 до 0.1), то превышение $\alpha'$ над $\alpha$ и $\beta'$ над $\beta$ пренебрежимо мало.

Последовательный критерий отношения правдоподобия (критерий Вальда) использует отношение функции правдоподобия для альтернативной гипотезы к функции правдоподобия для нулевой гипотезы, вычесленные на каждом шаге для растущей выборки из ГС.

$$
Z[X_{[i]}] = \frac{L(X_{[i]}, F_1)}{L(X_{[i]}, F_0)}
$$

По ЦПР смесь выборок из нормальных СВ стремится к нормальному распределению. Поэтому будем использовать функцию плотности вероятностей для нормального закона распределения.

Для нормально распределенной выборки при неизвестном мат ожидании и известном стандартном отклонении функции правдоподобия для нулевой и альтернативной гипотез можно вычислить по следующим формулам:

$$
L_0 (X_{[n]}) = \frac{1}{(2\pi)^{n/2} \sigma^n} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n (X_i - a_0)^2}
$$

$$
L_1 (X_{[n]}) = \frac{1}{(2\pi)^{n/2} \sigma^n} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n (X_i - a_1)^2}
$$

### Реализация критерия Вальда

````python
# Функция для проведения экспериментов с критерием Вальда
def vald(means, std, n, parts, a0, a1):
    # задаем 2 нормальных СВ
    # мат ожидания разные, стандартное отклонение одинаковое
    rvs = [st.norm(means[0], std), st.norm(means[1], std)]
    
    # выбираем из каждого распределения количество значений, соответствующих указанной доле
    samples = [rv.rvs(int(np.floor(part*n))) for part, rv in zip(parts, rvs)]
    sample = np.hstack(samples)
    
    sigma = np.std(sample)

    
    z = []
    subsample = []

    while(True):
        subsample.append(np.random.choice(sample, 1))
    
        n = len(subsample)
        likelihoods = []
        for a in (a0, a1):
            summ = sum([(x_i - a)**2 for x_i in subsample])
            likelihoods.append((1/(((2*math.pi)**(n/2)) * std**n)) * (math.exp((-1/(2*(std**2)))*summ)))
        try:
            z.append(likelihoods[1]/likelihoods[0])
        except:
            print('random value turned to division by 0, please, try again')
            break
        if((z[-1] < c0) | (z[-1] > c1)):
            break
    return z
````

### Нижняя оценка среднего числа наблюдений

Нижняя оценка среднего числа наблюдений для любого последовательного критерия при заданных $(\alpha, \beta)$ в случае справедливости нулевой и альтернативной гипотез:

$$
E_0\nu (\alpha, \beta) \geqslant -\frac{\omega(\alpha, \beta)}{E_{H_0}}
$$

$$
E_1\nu (\alpha, \beta) \geqslant -\frac{\omega(\alpha, \beta)}{E_{H_1}}
$$

где $\nu$ - номер шага, на котором алгоритм завершил работу

$\omega(\alpha, \beta)$ - некоторая функция

$$
\omega(x, y) = ((1 - x) ln (\frac{1-x}{y}) + x ln(\frac{x}{1-y}))
$$

### Требуемые формулы для расчета коэффициента корреляции

Выборочный коэффициент корреляции $r$ :

$$
r = \frac{\sum_{i=1}^n (X_i - \bar{X}) (Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^n (X_i - \bar{X})^2} {\sqrt{\sum_{i=1}^n (Y_i - \bar{Y})^2}}}
$$

где оценки среднего равны:

$$
    \bar{x} = \frac{1}{n}\sum_{i=1}^n x_i
$$

$$
    \bar{y} = \frac{1}{n}\sum_{i=1}^n y_i
$$

Распределение величены $r$ может быть преобразовано к виду:

$$
t = \frac{r}{\sqrt{1 - {r}^2}}\sqrt{n - 2} \sim t_{n-2}
$$

Итак, выборочное распределение этой статистики есть распределение Стьюдента с $n-2$ степенями свободы.

````python
def pearson_corr_coef(xs, ys):
    x_hat = 1/len(xs) * sum(xs)
    y_hat = 1/len(ys) * sum(ys)
    
    # сумма в числителе
    sum_numerator = 0
    for i in range(len(xs)):
        delta_x = xs[i] - x_hat
        delta_y = ys[i] - y_hat
        sum_numerator += delta_x * delta_y
    
    denom1 = math.sqrt(sum([(x_i - x_hat)**2 for x_i in xs]))
    denom2 = math.sqrt(sum([(y_i - y_hat)**2 for y_i in ys]))
    
    r = sum_numerator/(denom1*denom2)
    return r
````

````python
def student_t(r, n):
    t = r/(math.sqrt(1-r**2)) * math.sqrt(n-2)
    return t
````

### Подсчет коэффициента корреляции и проверка гипотезы о его значении

Примем нулевую гипотезу о том, что коэффициент корреляции равен 0. (Соответственно, по альтернативной гипотезе он не будет равен 0)

$H_0: r = 0$

````python
def count_corr_coef(df):
    r_FSIQ = pearson_corr_coef(df.MRI_Count, df.FSIQ)
    r_VIQ = pearson_corr_coef(df.MRI_Count, df.VIQ)
    r_PIQ = pearson_corr_coef(df.MRI_Count, df.PIQ)

    n = df.shape[0]

    t_FSIQ = student_t(r_FSIQ, n)
    t_VIQ = student_t(r_VIQ, n)
    t_PIQ = student_t(r_PIQ, n)

    print('Связь размера мозга с каждым из трех показателей уровня интеллекта:')
    print('FSIQ: r =', r_FSIQ, ', t =', t_FSIQ)
    print('VIQ: r =', r_VIQ, ', t =', t_VIQ)
    print('PIQ: r =', r_PIQ, ', t =', t_PIQ)

    print(str.format('\nКритические значения параметра t для распределения с {0} степенями свободы при уровне значимости равном 0.05',
                 n-2))
    print(st.t(n-2).ppf([0.025, 0.975]))
````

````python
count_corr_coef(df)
````

Во всех трех случаях значения критерия t превышают критическое значение для уровня значимости 0.05. Следовательно, можно отклонить нулевую гипотезу и принять значения коэффициента корреляции для совместной выборки мужчин и женщин:

#### Таблица сопряженности (с абсолютными частотами признаков) для столбцов Sex и PreferCar

Таблицы сопряженности используются для проверки гипотезы о наличии связи между двумя признаками (критерий хи-квадрат).

На пересечении строки и столбца указывается частота совместного появления fij соответствующих значений двух признаков xi и yj.

````python
sex1_list = []
sex2_list = []

for i in range(df.Sex.value_counts().count()):
    for j in range(df.PreferCar.value_counts().count()):
        nij = df[(df.Sex == i+1) & (df.PreferCar == j+1)].count()[0]
        if(i == 0):
            sex1_list.append(nij)
        else:
            sex2_list.append(nij)

crosstable = {'1': sex1_list, '2': sex2_list}
df_crosstab = pd.DataFrame(data=crosstable)
df_crosstab
````

````python
# подпишем строки и столбцы в таблице сопряженности
df_crosstab.columns.name = 'Sex'
df_crosstab.index = range(1, 5)
df_crosstab.index.name = 'PreferCar'
df_crosstab
````

#### Критерий хи-квадрат

Статистика $X^2$ :

$$
X^2 = n (\sum_{i=1}^K \sum_{j=1}^L \frac{n_{ij}^2}{n_i n_j} - 1) \sim \chi_{(K-1)(L-1)}^2
$$

K - количество строк в таблице сопряженности, L - количество столбцов в таблице сопряженности

````python
summ = 0
for i in range(K):
    for j in range(L):
        summ += (df_crosstab.iloc[i,j]**2)/(ni[i]*nj[j])
X2 = n*(summ - 1)
X2
````

$$\chi_{(K-1)(L-1)}^2 = \chi_{(4-1)(2-1)}^2 = \chi_3^2$$
